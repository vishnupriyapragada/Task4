```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_7_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_7_na_na_na&as-pos=3&as-type=RECENT&suggestionId=mobiles&requestId=7d1de872-4bfb-4cb1-bde2-ad006c4cf1ea&as-backfill=on"
```

```python
# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
#HTTP Request
webpage = requests.get(URL,headers=HEADERS)
webpage
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
soup
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'CGtC98'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
product_list = "https://flipkart.com" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, "html.parser") 
```

```python
new_soup
```

```python

name = new_soup.find("span", attrs={"class":'VU-ZEz'}).text.strip()
name
```

```python
price = new_soup.find("div", attrs={"class": "Nx9bqj CxhGGd"}).text.strip()
price
```
